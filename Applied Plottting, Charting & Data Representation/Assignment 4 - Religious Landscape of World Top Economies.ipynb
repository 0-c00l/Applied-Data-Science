```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import mpl, plt
import re
import seaborn as sns

%matplotlib inline
```

```python
relig = pd.read_csv(r'C:\Users\mhpel\Desktop\Data Science in Python\Applied Data Science with Python\Applied Plotting, Charting & Data Representation in Python\Week 4\Global Religion Data Analysis\WRP_national.csv')
```

```python
pct = relig[['year', 'name']].join(relig.loc[:, 'chrstprotpct':'total'])
```

```python
top_20_gdp = ['USA', 'CHN', 'JPN', 'GER', 'IND', 'FRA', 'UKG', 'ITA', 'BRA', 'CAN','ROK','RUS', 'AUS', 
              'SPN', 'MEX', 'INS','NTH', 'SAU','SWZ','TAW']
```

```python
mpl.rcParams['font.family'] = 'serif'
```

```python
total_pct = pct[[name for name in pct.columns.tolist() if len(re.findall('.*genpct', name))>0]]
```

```python
sample_2 = relig[['name']].join(total_pct).join(relig[['nonreligpct']])
```

```python
top_econ = round(sample_2[sample_2['name'].isin(top_20_gdp)],2)
```

```python
top_econ.columns = ['Country','Christianity', 'Judaism', 'Islam', 'Buddhism', 'Zoroastrian', 'Hindu', 'Sikh','Shinto', 'Baha`i','Taoism',
                  'Confucianism', 'Jain', 'Syncretic Religions', 'Animist Religions', 'Other Religions', 'Non. Religious']
```

```python
plt.figure(figsize = (14,14))

sns.heatmap(round(top_econ.groupby('Country').mean().transpose(),2), cmap = 'YlGn',
            linewidth = 1.5, linecolor = 'white', annot = True, cbar_kws = {'label':'% of Total Population', 
                                                                            'orientation':'horizontal'})

ax = plt.gca()
plt.xticks(rotation =0)
plt.yticks(rotation = 0, size = 12)
plt.xlabel('Country', size = 12, labelpad = 14)
plt.ylabel('Religion', size = 12)
plt.title('Religious Landscape of World Top 20 Economies', size = 15).set_position([.5,1.04])

plt.savefig('Religion', frameon = True)
```
